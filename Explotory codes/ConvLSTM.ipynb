```python
import torch

pretrained_checkpoint_path = "/home/dhu/ECE-228-Project/workspace/1000_iterations.pth"
checkpoint = torch.load(pretrained_checkpoint_path)["model"]

print(checkpoint.keys())
print(len(checkpoint.keys()))
for k, v in checkpoint.items():
  print(k, v.shape)
```

```python
from collections import OrderedDict
new_checkpoint = OrderedDict()

for k, v in checkpoint.items():
    if "base" in k:
        new_checkpoint[k.replace("base.", "")] = v
    elif "fc_transfer" in k:
        new_checkpoint[k.replace("fc_transfer", "fc2")] = v

for k, v in new_checkpoint.items():
    print(k, v.shape)
```
